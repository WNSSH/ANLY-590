### 2.

```python
t = 0.025
x = np.arange(-10.0, 10.0, t)
y = np.arange(-10.0, 10.0, t)
X, Y = np.meshgrid(x, y)
Z = 1 - X**3 + 100*(Y**2 - X)**2
fig, ax = plt.subplots()
plot2 = ax.contour(X, Y, Z)
```

### 3.

```python
#first define the derivatives

def grad1(vec):
    x,y=vec
    partial_x = -3*x**2 - 200*y**2 + 200*x
    partial_y = 400*(y**3 - x*y)
    return np.array([partial_x, partial_y])

#Then build the algorithm 

def grad_descent(x,y,iterations=10,lr=13):
    point = np.array([x,y])
    lines = [point]
    
    for i in range(iterations):
        temp = grad1(point)
        point = point - lr * temp
        lines.append(point)
    return np.array(lines)
```

```python
np.random.seed(314)
test=grad_descent(1,0,iterations=200,lr=0.01)
fig, ax = plt.subplots()
curve = ax.contour(X, Y, Z)
x= test[:,0]
y= test[:,1]
plt.plot(x,y,'-o')
```

### 4.

```python
def gdm(x=1,y=0,iterations=200,epsilon=1, alpha=1):
    point = np.array([x,y])
    lines = [point]
    v = np.zeros(point.size)
    
    for i in range(iterations):
        grad = grad1(point)
        v = alpha*v + epsilon*grad
        point = point - v
        lines.append(point)
    print("epsilon="+str(epsilon)+",alpha="+str(alpha))
    return np.array(lines)
```

### 1.

The loss function is 
$$L=-y_i log(\hat y_i) -( 1 - y_i ) log (1 - \hat y_i)$$

The derive expressions are: $$
\left\{
\begin{aligned}
\frac{\partial L}{\partial y_{out}} = -(\frac{y_i}{{\hat y}_i} - \frac{1 - y_i}{1 - {\hat y}_i}) \\
\frac{\partial y_{out}}{\partial y_{in}} = \sigma(y_{in}) (1 - \sigma(y_{in}))\\
\frac{\partial L}{\partial c} =\frac{\partial L}{\partial y_{out}} \frac{\partial y_{out}}{\partial y_{in}}\\
\frac{\partial L}{\partial v_i} = \frac{\partial L}{\partial y_{out}} \frac{\partial y_{out}}{\partial y_{in}} h_{out}\\
\frac{\partial \hat h_{out}}{\partial h_{in}} = 1(h>0),\space 0(else)\\
\frac{\partial L}{\partial w_{ij}} =   -(\frac{y_i}{{\hat y}_i} - \frac{1 - y_i}{1 - {\hat y}_i})  \sigma(y_{in}) (1 - \sigma(y_{in})) v \frac{\partial h_{out}}{\partial h_{in}} x_{i}
\end{aligned}
\right.$$

```python
#Functions
def relu(x):
    return np.maximum(x,0)
def sigmoid(x):
    return 1/(1+np.exp(-x))
def relu_d(x):
    return np.where(x>0, 1, 0)
def sigmoid_d(x):
    return sigmoid(x) * (1 - sigmoid(x))
def loss_1(y, y_hat):
    temp = y * np.log(y_hat)
    return  -temp.sum()
def loss_2(y, y_hat):
    return -(y/y_hat - (1-y)/(1-y_hat))
```

```python
#Parameters
def para_in_vec(x):
    w_11, w_12, w_13, w_21, w_22, w_23, b_1, b_2, b_3, v_11, v_12, v_13, c =x
    w = np.array([[w_11,w_12,w_13],[w_21,w_22,w_23]])
    b = np.array([b_1, b_2, b_3])
    v = np.array([v_11, v_12, v_13])
    c = c
    return(w, b, v, c)

def ind_para(w, b, v, c):
    w_11, w_12, w_13, w_21, w_22, w_23 = w[0,0], w[0,1],w[0,2], w[1,0], w[1,1], w[1,2]
    b_1, b_2, b_3 = b[0], b[1], b[2]
    v_11, v_12, v_13 = v[0], v[1], v[2]
    c = c
    return np.array([w_11, w_12, w_13, w_21, w_22, w_23, b_1, b_2, b_3, v_11, v_12, v_13, c])
def yhat(x, w, b, v, c):
    temp1 = x.dot(w) + b
    hidden = relu(temp1)
    temp2 = hidden.dot(v) + c
    y_hat = sigmoid(temp2)
    return y_hat

def grad_para(para, x, y):
    w, b, v, c=para_in_vec(para)
    #print(v)
    v_new = np.reshape(v, (1,3))
    #print(v_new)
    temp1 = x.dot(w) + b
    hidden = relu(temp1)
    temp2 = hidden.dot(v) + c
    y_hat = sigmoid(temp2)
    y_sigmoid = loss_2(y, y_hat) * sigmoid_d(temp2)
    y_sigmoid_new = np.reshape(y_sigmoid,(500,1))
    
    d_v = hidden.T.dot(y_sigmoid_new)
    #print(d_v)
    #print(d_v.shape)
    #print(d_v[1][0])
    d_v=np.array([d_v[0][0],d_v[1][0],d_v[2][0]])
    #print("vvv")
    #print(d_v)
    d_c = y_sigmoid.sum(axis=0)
    #print(d_c)
    #print(d_c.shape)
    d_b = (y_sigmoid_new.dot(v_new)*(hidden > 0)).sum(axis=0)
    #print(d_b)
    #print(d_b.shape)
    d_w = x.T.dot(y_sigmoid_new.dot(v_new)*(hidden > 0))
    #print(d_w)
    #print(d_w.shape)
    para = ind_para(d_w, d_b, d_v, d_c)
    #print(para)
   
    return np.array(para)
```

```python
def grad_f(x, y,  iterations=10, lr=0.01):
    
    point = np.random.uniform(-.1,.1,size=13)
    lines = [point]
    losses = [loss_1(y, yhat(x, *para_in_vec(point)))]
    
    for i in range(iterations):
        grad = grad_para(point, x, y)
        #print(grad.shape)
        point = point - lr * grad
        #print(point.shape)
        
        lines.append(point)
        losses.append(loss_1(y, yhat(x, *para_in_vec(point))))
    return (np.array(lines), losses)
```

```python
def grad_m_f(x, y,  iterations=2000, alpha=.1, epsilon=1):
    point = np.random.uniform(-.1,.1,size=13)
    lines = [point]
    losses = [loss_1(y, yhat(x, *para_in_vec(point)))]
    v = np.zeros(point.size)
    
    for i in range(iterations):
        grad = grad_para(point, x, y)
        v = alpha*v + epsilon*grad
        point = point -  v
        lines.append(point)
        losses.append(loss_1(y, yhat(x, *para_in_vec(point))))
    return (np.array(lines), losses)
    
```
